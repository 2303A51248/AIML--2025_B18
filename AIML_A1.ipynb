```python
import numpy as np


list1 = [1, 2, 3, 4, 5]
list2 = [6, 7, 8, 9, 10]


array1 = np.array(list1)
array2 = np.array(list2)

print(type(list1))
print(type(array1))


array_mul = array1 * array2
array_div = array1 / array2
array_pow = array1 ** 2


print(f"Addition of Two: {array1} + {array2} = {array1 + array2}")


array_sin = np.sin(array1)
array_log = np.log(array1)
array_log2 = np.log2(array1)
array_exp = np.exp(array1)


print("Sin:", array_sin)
print("Log:", array_log)
print("Log2:", array_log2)
print("Exp:", array_exp)
```

```python
import matplotlib.pyplot as plt


people = ['Kiran', 'Arun', 'Vijay', 'Varun']
age = [25, 30, 35, 40]
height = [145, 151, 165, 173]
weight = [45, 55, 65, 75]


plt.figure(figsize=(10, 5))
plt.scatter(height, weight, color='blue', marker='o')
plt.title('Height vs Weight Scatter Plot')
plt.xlabel('Height (cm)')
plt.ylabel('Weight (kg)')
plt.grid(True)
plt.show()

x = range(len(people))
plt.figure(figsize=(12, 6))
plt.bar(x, height, width=0.4, label='Height', color='orange', align='center')
plt.bar([i + 0.4 for i in x], weight, width=0.4, label='Weight', color='green', align='center')

plt.title('Height and Weight Bar Chart')
plt.xlabel('People')
plt.ylabel('Measurement (cm for Height, kg for Weight)')
plt.xticks([i + 0.2 for i in x], people)
plt.legend()
plt.show()

plt.figure(figsize=(12, 15))


plt.subplot(3, 1, 1)
plt.hist(age, bins=4, color='blue', edgecolor='black')
plt.title('Age Distribution')
plt.xlabel('Age (years)')
plt.ylabel('Frequency')


plt.subplot(3, 1, 2)
plt.hist(height, bins=4, color='green', edgecolor='black')
plt.title('Height Distribution')
plt.xlabel('Height (cm)')
plt.ylabel('Frequency')


plt.subplot(3, 1, 3)
plt.hist(weight, bins=4, color='red', edgecolor='black')
plt.title('Weight Distribution')
plt.xlabel('Weight (kg)')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```
